```python
import json
import pandas as pd
import pickle
```

```python
DICTIONARY_DATA_DIR = 'Dictionary_data/'
COMPLETE_WORDLIST = set()
```

```python
def load_csv(file_name, pd_args=None):
    encodings_to_try = ['utf-8', 'Latin-1', 'ISO-8859-1']
    
    for encoding in encodings_to_try:
        try:
            df = pd.read_csv(DICTIONARY_DATA_DIR + file_name, encoding=encoding, **(pd_args or {}))
            print("File read successfully with encoding:", encoding)
            print(df.head())
            return df
        except UnicodeDecodeError:
            pass
    raise ValueError("Failed to read the file with the provided encodings.")
```

```python
# https://www.kaggle.com/datasets/rafsunahmad/popular-quotes-author-classifier?resource=download
file_name = 'Autor_detection.csv'
df_author = load_csv(file_name)
writers = set(df_author.loc[:]['Writter name'])
print(len(writers))
COMPLETE_WORDLIST.update(writers)
```

```python
# https://www.kaggle.com/datasets/joebeachcapital/art-history
file_name = 'artists.csv'
df_artists = load_csv(file_name)
artists = set(df_artists.loc[:]['artist_name'])
print(len(artists))
COMPLETE_WORDLIST.update(artists)
```

```python
# https://www.kaggle.com/datasets/rishidamarla/art-and-artists-from-the-museum-of-modern-art?select=Artworks.csv
file_name = 'Artworks.csv'
df_artworks = load_csv(file_name)
artworks = set(df_artworks.loc[:]['Artist'])
print(len(artworks))
COMPLETE_WORDLIST.update(artists)
```

```python
# https://www.kaggle.com/datasets/mfrancis23/museum-of-modern-art-collection?select=Artists.csv
file_name = 'Artists.csv'
df_moma_artists = load_csv(file_name)
moma_artists = set(df_moma_artists.loc[:]['DisplayName'])
print(len(moma_artists))
COMPLETE_WORDLIST.update(moma_artists)
```

```python
# https://www.kaggle.com/datasets/anthonytherrien/dictionary-of-english-words-and-definitions
file_name = 'dict.csv'
df_dictionary = load_csv(file_name)
dictionary_words = set(df_dictionary.loc[:]['word'])
print(len(dictionary_words))
COMPLETE_WORDLIST.update(dictionary_words)
```

```python
# https://www.kaggle.com/datasets/rtatman/dictionary-of-american-regional-english-dareds
file_name = 'geodare.json'
json_geodare = json.load(open(DICTIONARY_DATA_DIR + file_name, 'r', encoding='utf-8-sig'))
df_geodare = pd.DataFrame(json_geodare)
geodare_words = set(df_geodare['word'])
print(len(geodare_words))
COMPLETE_WORDLIST.update(geodare_words)
```

```python
# https://www.kaggle.com/datasets/yetman/english-contractions
file_name = 'contractions.json'
json_contractions = json.load(open(DICTIONARY_DATA_DIR + file_name, 'r', encoding='utf-8-sig'))
contractions = set(json_contractions.keys())
print(len(contractions))
COMPLETE_WORDLIST.update(contractions)
```

```python
# https://www.kaggle.com/datasets/dfydata/the-online-plain-text-english-dictionary-opted
file_name = 'OPTED-Dictionary.csv'
df_opted = load_csv(file_name)
opted = set(df_opted.loc[:]['Word'])
print(len(opted))
COMPLETE_WORDLIST.update(opted)
```

```python
# https://www.kaggle.com/datasets/bwandowando/479k-english-words?select=words.txt
file_name = 'words.txt'
wordlist = set([word for word in open(DICTIONARY_DATA_DIR + file_name, 'r', encoding='utf-8-sig').read().splitlines()])
print(len(set(wordlist)))
COMPLETE_WORDLIST.update(wordlist)
```

```python
print("Total unique words collected:", len(COMPLETE_WORDLIST))
```

```python
with open(DICTIONARY_DATA_DIR + 'complete_wordlist.pkl', 'wb') as f:
    pickle.dump(COMPLETE_WORDLIST, f)
```
